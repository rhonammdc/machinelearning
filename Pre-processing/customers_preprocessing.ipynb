# Customers Data Preprocessing

## Import Required Libraries

```
import pandas as pd
import numpy as np
print("Libraries imported successfully")
```

## Load Raw Data

```
# Load customers data
customers = pd.read_csv("customers_data.csv")
print("Original Customers Data Shape:", customers.shape)
```

## Data Inspection

```
# Display first few rows
print("First few rows of the data:")
print(customers.head())
```

```
# Check data types
print("Data Types:")
print(customers.dtypes)
```

```
# Check missing values
print("Missing Values:")
print(customers.isnull().sum())
```

```
# Check basic statistics
print("Basic Statistics:")
print(customers.describe())
```

## Data Cleaning

```
# Create copy for cleaning
customers_cleaned = customers.copy()
```

```
# Handle missing Company_ID
customers_cleaned["Company_ID"] = customers_cleaned["Company_ID"].fillna(customers_cleaned["Company_ID"].max() + 1)
print("Missing Company_IDs handled")
```

```
# Handle missing Company_Profit
customers_cleaned["Company_Profit"] = customers_cleaned["Company_Profit"].fillna(customers_cleaned["Company_Profit"].median())
print("Missing Company_Profit handled")
```

## Verification

```
# Check for any remaining missing values
print("Missing Values After Cleaning:")
print(customers_cleaned.isnull().sum())
```

```
# Display cleaned data statistics
print("Cleaned Data Statistics:")
print(customers_cleaned.describe())
```

## Save Cleaned Data

```
# Save cleaned customers data
customers_cleaned.to_csv("customers_cleaned.csv", index=False)
print("Cleaned customers data saved to customers_cleaned.csv")
```
